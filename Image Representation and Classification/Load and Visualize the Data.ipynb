# Day and Night Image Classifier

The day/night image dataset consists of 200 RGB color images in two categories: day and night. There are equal numbers of each example: 100 day images and 100 night images.

Goal: Build a classifier that can accurately label these images as day or night, and that relies on finding distinguishing features between two types of images.

*Note: All images come from the [AMOS dataset](http://cs.uky.edu/~jacobs/datasets/amos/) (Archive of Many Outdoor Scenes).*

### Import Resources

```python
import cv2 # computer vision library
import helpers

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

### Traning and Testing Data

The 200 day/night images are separated into training and testing datasets.
* 60% of these images are training images, to create a classifier.
* 40% are test images, which will be used to test the accuracy of the classifier.

```python
# Image data directories
image_dir_training = "day_night_images/training/"
image_dir_testing = "day_night_images/test/"
```

### Load the Datasets
Load the training day/night images and store all of them in a variable, `IMAGE_LIST`. This list contains the images and their associated label ("day" or "night")

The first image-label pair can be accessed by index: 
`IMAGE_LIST[0][:]`

```python
# Using the load_dataset function in helpers.py
# Load training data
IMAGE_LIST = helpers.load_dataset(image_dir_training)
```

### 1. Visualize the Input Images

```python
# Select an image and its label by list index
image_index_day = 0
selected_image = IMAGE_LIST[image_index_day][0]
selected_label = IMAGE_LIST[image_index_day][1]

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
# Print out the shape of the image and The image's label 'selected_label'
ax1.set_title('Day')
ax1.imshow(selected_image)
print(selected_image.shape)
print(selected_label)

# Display a night image (Nigt images have index number > 119)
image_index_night = 120
selected_image_new = IMAGE_LIST[image_index_night][0]
selected_label_new = IMAGE_LIST[image_index_night][1]

ax2.set_title('Night')
ax2.imshow(selected_image_new)
print("----------------------------")
print(selected_image_new.shape)
print(selected_label_new)
```
